```python
class A:
    def method_0(self):
        return "class A"
    def method_2(self):
        return "class A"

class B(A):
    def method_0(self):
        return super().method_0()
    def method_1(self):
        return "class B"

class C(A):
    def method_0(self):
        return "class C"
    def method_1(self):
        return "class C"
    def method_2(self):
        return "class C"

class D(B, C):
    pass

obj = D()

print(obj.method_0(), obj.method_1(), obj.method_2())
```
