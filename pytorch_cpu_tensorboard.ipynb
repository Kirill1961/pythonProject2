```python
import torch
import time
from torch.utils.tensorboard import SummaryWriter

```

```python
# ЯВНО указываем CPU
device = torch.device("cpu")

# Создаём тензор
x = torch.tensor([1.0, 2.0, 3.0]).to(device)

# Простая операция
y = x * 2

print("x:", x)
print("y:", y)
print("Устройство x:", x.device)
print("Устройство y:", y.device)
```

```python


device = torch.device("cpu")

x = torch.randn(3000, 3000).to(device)
start = time.time()

y = torch.matmul(x, x)

end = time.time()

print("Устройство:", y.device)
print("Время (сек):", round(end - start, 3))
```

```python
# TODO Папка с логами
writer = SummaryWriter(log_dir="runs/exp1")
```

```python
for epoch in range(10):
    train_loss = 1 / (epoch + 1)
    val_loss = train_loss + 0.1

    writer.add_scalar("Loss/train", train_loss, epoch)
    writer.add_scalar("Loss/val", val_loss, epoch)
```

```python
writer.close()
```
